```python
from apng import APNG, PNG, FrameControl
import base64
import cv2
import numpy as np
import matplotlib.pyplot as plt
import io

## memory buf usage:
# https://stackoverflow.com/questions/52865771/write-opencv-image-in-memory-to-bytesio-or-tempfile
```

```python
imgs = [np.full([100,100,4], 255, dtype=np.uint8) for _ in range(100)]
for i in range(len(imgs)):
    imgs[i][i:i+10, i:i+10] = [0,0,0,255]

# bufs = [io.BytesIO(cv2.imencode('.png', img)[1]) for img in imgs]
# pngs = [PNG.from_bytes(buf.read()) for buf in bufs]

pngs = [PNG.from_bytes(cv2.imencode('.png', img)[1].tobytes()) for img in imgs]
```

```python
apng = APNG()
fps = 50
for i in pngs:
    apng.append(i, delay=1000//fps)
apng.save('out.png')
```

```python
src = list(bytearray(apng.to_bytes()))
```
